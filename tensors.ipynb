```python
import time 
```

```python
device_gpu = torch.device('cuda')
device_cpu = torch.device('cpu')
print(device_gpu)
print(device_cpu)
```

```python
# creating a tensor on gpu 
torch.rand((2,3),device=device_gpu)
```

```python
# creating a tensor on cpu
a_cpu = torch.rand((2,3),device=device_cpu)
```

```python
# moving existing tensor on gpu
a_cpu.to(device=device_gpu)
a_cpu + 5
```

```python
#  Comparison of cpu and gpu 
size = 10000

# create random matrix on cpu
matrix_cpu_1 = torch.rand(size,size)
matrix_cpu_2 = torch.rand(size,size)


#  time measure on cpu
start_time = time.time()
result_cpu = torch.matmul(matrix_cpu_1,matrix_cpu_2)

cpu_time = time.time() - start_time

print(f"Time on cpu : {cpu_time: .4f} seconds")

# Move Matrix on gpu 

matrix_gpu_1 = matrix_cpu_1.to('cuda')
matrix_gpu_2 = matrix_cpu_2.to('cuda')

#  time measure on gpu
start_time = time.time()
result_gpu = torch.matmul(matrix_gpu_1,matrix_gpu_2)
torch.cuda.synchronize()  # Ensure all GPU operations are complete 

gpu_time = time.time() - start_time

print(f"Time on gpu : {gpu_time: .4f} seconds")
```
